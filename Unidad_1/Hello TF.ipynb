```python
ds = tf.data.Dataset.from_tensor_slices((tf.random.uniform([4,3]),
                                    tf.range(4)))
```

```python
for ex in ds:
    print('X:', ex[0].numpy(),'y:',ex[1].numpy())
```

```python
ds = ds.shuffle(buffer_size=len(ds))
for ex in ds:
    print('X:', ex[0].numpy(),'y:',ex[1].numpy())
```

```python
batch = ds.batch(batch_size=3,drop_remainder=False)
b_x,b_y = next(iter(batch))

print('Batch x:',b_x.numpy())
print('Batch y:',b_y.numpy())
```

```python
dsx = ds.shuffle(buffer_size=len(ds)).batch(2).repeat(count=4)
for ex in dsx:
    print('X:', ex[0].numpy(),'y:',ex[1].numpy())
```

```python
import os
import matplotlib.pyplot as plt

ruta = 'ucl_logos/'
l_arch = sorted(os.listdir(ruta))
```

```python
len(l_arch),l_arch
```

```python
fig = plt.figure(figsize=(16,16))
for i,arch in enumerate(l_arch,1):
    print(arch)
    img = tf.image.decode_image(tf.io.read_file(os.path.join(ruta,arch)))
    ax = fig.add_subplot(4,4,i)
    ax.set_xticks([])
    ax.set_yticks([])
    ax.imshow(img)
    ax.set_title(arch)
plt.tight_layout()
plt.show()
```

```python
etiq = tf.data.Dataset.from_tensor_slices((list(map(lambda x:os.path.join(ruta,x),l_arch)),
                                           [x[:-4] for x in l_arch]))

for item in etiq:
    print(item[0].numpy(),item[1].numpy())
```

```python
def carga_y_pre(ruta,etiq):
    imag = tf.io.read_file(ruta)
    imag = tf.image.decode_png(imag,channels=3)
    imag = tf.image.resize(imag,[imh,imw])
    imag/= 255.0
    return imag, etiq

imh, imw = 500,500

ds = etiq.map(carga_y_pre)

fig = plt.figure(figsize=(16,16))
for i,im in enumerate(ds,1):
    ax = fig.add_subplot(4,4,i)
    ax.set_xticks([])
    ax.set_yticks([])
    ax.imshow(im[0])
    ax.set_title("%s"%im[1].numpy(),size=20)
plt.tight_layout()
plt.show()
```

```python
mnist,mnist_info = tfds.load('mnist',with_info=True,
                             shuffle_files=True)
#print(mnist_info)

Xt = mnist['train']
Xt = Xt.map(lambda x:(x['image'],x['label']))
Xt = Xt.batch(10)
batch = next(iter(Xt))

print(batch[0].shape,batch[1])
```

```python
fig = plt.figure(figsize=(15,6))

for i,(img,etiq) in enumerate(zip(batch[0],batch[1]),1):
    ax = fig.add_subplot(2,5,i)
    ax.set_xticks([]) ; ax.set_yticks([])
    ax.imshow(img,cmap='gray_r')
    ax.set_title('%s'%etiq.numpy(),size=30)
plt.show()
```
